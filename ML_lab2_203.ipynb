**Entropy Function**

```python
def entropy(df):
    attr = df.keys()[-1]   
    entropy = 0
    values = df[attr].unique()
    for value in values:
        fraction = df[attr].value_counts()[value]/len(df[attr])
        entropy += -fraction*np.log2(fraction)
    return entropy
```

```python
def att_entropy(df,attribute):
  attr = df.keys()[-1]  
  target_variables = df[attr].unique()  
  variables = df[attribute].unique()    
  entropy2 = 0
  for variable in variables:
      entropy = 0
      for target_variable in target_variables:
          num = len(df[attribute][df[attribute]==variable][df[attr] ==target_variable])
          den = len(df[attribute][df[attribute]==variable])
          fraction = num/(den+eps)
          entropy += -fraction*log(fraction+eps)
      fraction2 = den/len(df)
      entropy2 += -fraction2*entropy
```

```python
def max_attr(df):
    Entropy_att = []
    IG = []
    for key in df.keys()[:-1]:

        IG.append(entropy(df)-att_entropy(df,key))
    return df.keys()[:-1][np.argmax(IG)]
```

```python
def attr_table(df, node,value):
  return df[df[node] == value].reset_index(drop=True)
```

**Building Tree**

```python
def buildTree(df,tree=None): 
    attr = df.keys()[-1]   
    node = max_attr(df)
   
    attValue = np.unique(df[node])
    
    if tree is None:                    
        tree={}
        tree[node] = {}
    

    for value in attValue:
        
        subtable = attr_table(df,node,value)
        clValue,counts = np.unique(subtable['play'],return_counts=True)                        
        
        if len(counts)==1:
            tree[node][value] = clValue[0]                                                    
        else:        
            tree[node][value] = buildTree(subtable) 
                   
    return tree
```
