```python
def initalize(self,x):
  weights = np.zero((x.shape(x)[1]+1,1))
  x = np.c_[np.ones((x.shape(x)[0],1)),x]
  return weights,x
```

```python
def sigmoid(self,z):
  sig = 1/1+e**(-z)
  return sig
```

```python
def cost(theta):
  z = dot(x,theta)
  cost0 = y.T.dot(math.log(self.sigmoid(z)))
  cost1 = (1-y).T.dot(math.log(1-self.sigmoid(z)))

  cost = -((cost1+cost0))/len(y)
  return cost
```
